## append method

```python
dd1={"oop":{'ist':70},"cpp":{'2nd':80},"java":{'3rd':90},"Ds":{'4th':95}}
```

```python
df5=pd.DataFrame(dd1)
```

```python
df5['oop']=[79,40,90,80]
```

```python
df5['ist']=[79,40,90,80]
```

```python
df5=df5.drop(['ist'],axis=1)
```

```python
df5.copy()
```

```python
df5
```

```python
df5.loc['2nd','java']=[37]
```

```python
df5
```

## add row

```python
dd3={'oop':40,'cpp':70,'java':80,'Ds':90}
```

```python
df5
```

```python
df5=df5.append(dd3,ignore_index=True)
```
